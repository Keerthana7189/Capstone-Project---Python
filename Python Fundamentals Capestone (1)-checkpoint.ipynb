Task 2: Fill Missing Cost Values Using Running Average

```python
## Loading the project Data

df_project=pd.read_csv('project.csv')
df_project
```

```python
## Converting the cost column to list

costs=df_project['Cost'].tolist()
##Looping each element to fill the missing values
for i in range(len(costs)):
    if pd.isna(costs[i]):
        ##Taking all previous valid values
               previous_values=[x for x in costs[:i] if pd.notna(x)]
        ##Computing average if previous values exists
               if previous_values:
                  average=np.mean(previous_values)
                  costs[i]=average
##Updating DataFrame
df_project['Cost']=costs
df_project
```

```python
df_project.to_csv('project.csv',index=False) ##Saving the Data
```

Task 3: Split Name into First and Last Name

```python
## Loading Employee Data

df_employee=pd.read_csv('employee.csv')

## Splitting the Name into First and last Name
df_employee[['First Name','Last Name']]=df_employee['Name'].str.split(' ',n=1,expand=True)

## Removing the Old 'Name' column
df_employee.drop(columns='Name',inplace=True)

## View the updated employee table
df_employee
```

```python
df_employee.to_csv('employee.csv',index=False)
```

Task 4: Join All Three DataFrames

```python
## Loading all the three DataFrame

df_employee=pd.read_csv('employee.csv')
df_seniority=pd.read_csv('seniority.csv')
df_project=pd.read_csv('project.csv')
```

```python
## Merging employee and seniority data using the 'ID'

df_emp_seniority=pd.merge(df_employee,df_seniority,on='ID',how='left')

## Merging the results with Project data using 'ID'

final_df=pd.merge(df_project,df_emp_seniority,on='ID',how='left')

## Final merged table
final_df
```

```python
final_df.to_csv('final.csv',index=False)
```

Task 5: Add a Bonus Column for Finished Projects

```python
## Loading the final merged table

final_df=pd.read_csv('final.csv')
final_df
```

```python
## Adding the Bonus of 5% for the finished project

final_df['Bonus']=final_df.apply(
    lambda row: 0.05*row['Cost'] if row['Status']=='Finished' else 0,
    axis=1
)
final_df
```

```python
final_df.to_csv('final.csv',index=False) ## Saving the updated final
```

Task 6: Demote Designation Level for Failed Projects

```python
## Loading the final merged table

final_df=pd.read_csv('final.csv')
final_df
```

```python
## Getting list of employee IDs who had at least one failed Project

failed_ids=final_df[final_df['Status']=='Failed']['ID'].unique()
failed_ids
```

```python
## Loading seniority data

df_seniority=pd.read_csv('seniority.csv')

## Demote each employee by level 1 if they are in Failed list
for emp_id in failed_ids:
    df_seniority.loc[df_seniority['ID']==emp_id,'Designation Level']+=1
    df_seniority
```

```python
## Removing employees whose designation level becomes>4

df_seniority=df_seniority[df_seniority['Designation Level']<=4]
df_seniority
```

```python
df_seniority.to_csv('seniority.csv',index=False) ## Saving the updated Seniority Data
```

Task 7: Add Mr/Mrs. Prefix and drop Gender Column

```python
## Loading the employee Data

df_employee=pd.read_csv('employee.csv')
df_employee
```

```python
## Adding 'Mr' for Males and 'Mrs' for Females

df_employee['First Name']=df_employee.apply(
    lambda row: 'Mr.'+row['First Name'] if row['Gender']=='M' else 'Mrs.'+row['First Name'],
    axis=1
)
df_employee 
```

```python
## Removing the Gender Column

df_employee.drop(columns='Gender',inplace=True)
df_employee
```

```python
df_employee.to_csv('employee.csv',index=False)  ## Saving the updated file
```

Task 8: Promote Designation level

```python
## Loading employee and Seniority data

df_employee=pd.read_csv('employee.csv')
df_seniority=pd.read_csv('seniority.csv')
```

```python
## Finding IDs of employees whose age is greater than 29

promote_ids=df_employee[df_employee['Age']>29]['ID'].values
promote_ids
```

```python
## Promote by decreasing the designation level by 1

for emp_id in promote_ids:
    current_level=df_seniority.loc[df_seniority['ID']==emp_id,'Designation Level'].values[0]
    if current_level>1:  ## Only promotes if they are not already at the top level
        df_seniority.loc[df_seniority['ID']==emp_id,'Designation Level']=current_level-1
df_seniority
```

```python
df_seniority.to_csv('seniority.csv',index=False)  ## Saving the updated file
```

Task 9: Create a Total Project Cost per Employee

```python
final_df=pd.read_csv('final.csv')
```

```python
## Group by employee ID and summing the project costs

total_cost=final_df.groupby('ID')['Cost'].sum().reset_index()
total_cost.rename(columns={'Cost':'Total Cost'},inplace=True)
total_cost
```

```python
## Gets ID and First Name from final df by dropping duplicates

name_info=final_df[['ID','First Name']].drop_duplicates()

## Merging the total costs with First Name
df_total_project=pd.merge(total_cost,name_info,on='ID')

## Reorder columns
df_total_project=df_total_project[['ID','First Name','Total Cost']]
df_total_project
```

```python
df_total_project.to_csv('TotalProject.csv',index=False) ## Saving the updated file
```
